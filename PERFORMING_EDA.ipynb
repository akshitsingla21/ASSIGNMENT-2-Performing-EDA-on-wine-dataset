```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/wine.data")
df.head()

df=df.drop(['14.23'],axis=1)
df.head()
```

```python
df.shape
print("no. of duplicate rows are",df[df.duplicated()].shape)

#to drop duplicate rows
df = df.drop_duplicates()

```

```python
 #renaming the columns
 df = df.rename(columns={"1":"col1","1.71":"col2"})
 df.head()
```

```python
#dropping the missing or null values
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()

print(df.isnull().sum())
```
